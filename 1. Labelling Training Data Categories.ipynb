```python
#most likely:
a = (df.Mechanism == 1) & (df.BPmed == 1)
b =  (df.Rank == 1) & (df.IPABP == 1) & (df.IPAHTN == 1) & (df.SideeffectFreq==1) 
c = (df.BPmed == 1) & (df.Rank >= 1) #genes meeting this criteria also appeared in omim, ensembl and genie
d = (df.Catalog >= 1) & (df.Ensembl == 1) & (df.IPABP == 1) & (df.IPAHTN == 1)

#likely:
e = (df.BPmed == 1) & (df.Catalog >= 1)
f = (df.SideeffectFreq==1) & (df.Rank >= 1)
g = (df.BPmed == 1) & (df.category !='certain')
h = (df.qvalAT <= 0.05) & (df.qvalLV <= 0.05)  & (df.qvalleg >= 0.05)  & (df.qvalpancr >= 0.05) & (df.qvalspleen >= 0.05)   & (df.qvalhippo >= 0.05) 
i = (df.qvalBL <= 0.05) & (df.qvalhippo <= 0.05)  & (df.qvalleg >= 0.05) & (df.qvalpancr >= 0.05) & (df.qvalspleen >= 0.05)  & (df.qvalhippo >= 0.05)  
j = (df.qvalAD <= 0.05) & (df.qvalhippo >= 0.05)  & (df.qvalleg >= 0.05)  & (df.qvalpancr >= 0.05) & (df.qvalspleen >= 0.05)  & (df.qvalhippo >= 0.05)  
k = (df.qvalCO <= 0.05) &  (df.qvalAO <= 0.05) & (df.qvalTIB <= 0.05)  & (df.qvalleg >= 0.05)  & (df.qvalpancr >= 0.05) & (df.qvalspleen >= 0.05) & (df.qvalhippo >= 0.05)

#possible:
l = (df.Mechanism == 1)
m =  (df.Rank >= 1) & (df.category !='certain')
n = (df.Disease >= 1) &(df.qvaladiSub <= 0.05) & (df.qvaladiVsc <= 0.05)
o = (df.Disease >= 1) & (df.qvalpituit <= 0.05) 
p = (df.Disease >= 1) & (df.qvalthyroid <= 0.05) 
q = (df.SideeffectFreq==1)

#Least likely: 

r = (df.Disease == 0) & (df.IPABP == 0)  & (df.IPAHTN == 0) & (df.Catalog != 3)  & (df.Catalog != 2) & (df.qvalpancr <= 0.05) & (df.qvalspleen <= 0.05)  & (df.qvalhippo <= 0.05)  & (df.qvalleg <= 0.05) 
s = (df.Disease == 0) & (df.IPABP == 0)  & (df.IPAHTN == 0) & (df.Catalog != 3)  & (df.Catalog != 2) & (df.qvalstomach <= 0.05) & (df.qvalileum <= 0.05) & (df.qvalEsoph <= 0.05) & (df.qvalSaliv <= 0.05)



df['category'] = np.select([a, b, c, d, e, f, g, h, i, j, k, l, m, n, o, p, q, r, s], ['most likely','most likely','most likely','most likely', 
                                                                                'likely', 'likely', 'likely', 
                                                                                'likely', 'likely', 'likely', 'likely',
                                                                               'possible', 'possible','possible', 'possible', 'possible', 'possible',
                                                                              'least likely', 'least likely'], default='unknown')

print(df.category)
```
